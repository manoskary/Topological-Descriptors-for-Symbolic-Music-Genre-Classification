# A notebook to observe Self Similarity Matrices in Symbolic Music

In this notebook we use self similarity matrices to study possible segmentation techniques in Symbolic music. 
The need of a feature space that produces elements of the same size lead us to the use of interval vectors to represent chords. Therefore our feature space $L = (\mathbb{Z}/\mathbb{Z}_{12})^6$.

```python
import music21 as ms
import numpy as np
```

Extract chords and transform in interval Vectors

```python
m = ms.converter.parse("Midi_files/002_Keith_Jarrett.mid")
chords = m.chordify()

intVec = list()
for thisChord in chords.recurse().getElementsByClass('Chord'):
    intVec.append(thisChord.intervalVector)
```

```python
Create the Selfsimilarity matrix based on the Frobenious norm:
    $$ S = [ \sum_{i, j} abs(v_i \cdot v_j)^2 ]^{\frac{1}{2}},  $$
```

```python
M = np.array(intVec + intVec)
M = M.reshape(len(intVec), 2, 6)
S = np.zeros((len(intVec), len(intVec)))
for i in range(len(intVec)):
    for j in range(len(intVec)):
        S[i][j] = np.dot(intVec[i], intVec[j])
```

```python
def normalize(v):
    norm=np.linalg.norm(v, ord=1)
    if norm==0:
        norm=np.finfo(v.dtype).eps
    return v/norm

S = normalize(S)
```

Apply Mathematical Morphology Processes

```python
from matplotlib.pyplot import imshow

img = S.T
imshow(img, cmap='gist_gray')
```

```python
from skimage.morphology import erosion, dilation, binary_opening, closing
from skimage.morphology import disk, square
from skimage.segmentation import flood, flood_fill

img = S
img = binary_opening(img, square(6))
img = flood_fill(img, (10,10), 10)
img = closing(img, square(6))

imshow(img, cmap='gist_gray')
```

Apply k-means Clustering

```python
import matplotlib.pyplot as plt
# For clustering the image using k-means, we first need to convert it into a 2-dimensional array
image_2D = img.reshape(img.shape[0], img.shape[1])
# Use KMeans clustering algorithm from sklearn.cluster to cluster pixels in image
from sklearn.cluster import KMeans
# tweak the cluster size and see what happens to the Output
kmeans = KMeans(n_clusters=5, random_state=0).fit(image_2D)
clustered = kmeans.cluster_centers_[kmeans.labels_]
# Reshape back the image from 2D to 3D image
clustered_3D = clustered.reshape(img.shape[0], img.shape[1])
plt.imshow(clustered_3D)
plt.title('Clustered Image')
plt.show()
```
